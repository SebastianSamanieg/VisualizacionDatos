```python
df_tips = pd.read_csv('/content/tips.csv')
```

```python
df_tips.head(10)
```

```python
hue_colors = {"Yes": "red", "No": "green"}

sns.scatterplot(x="total_bill", y="tip", data=df_tips, hue="smoker", hue_order=["Yes", "No"], palette=hue_colors.values())
plt.show()
```

```python
sex_color = {"Female": "magenta", "Male": "blue"}

sns.countplot(x="smoker", data=df_tips, hue="sex", palette=sex_color)
plt.show()
```
