## Punto 2.2 Temperaturas Anuales 50%

En el archivo "temperaturas.csv" se listan las temperaturas por años de las ciudades de Bogotá y Tunja.
Realice un gráfico de líneas para comparar la evolución de la temperatura de estas dos ciudades.


```python
#Escriba su código aquí
df_temperatures = pd.read_csv('/content/temperaturas.csv')
df_temperatures.head(9)
```

```python
sns.lineplot(x='anios', y='bogota',label='Bogota', data = df_temperatures)
sns.lineplot(x='anios', y='tunja', label='Tunja',data = df_temperatures)
plt.ylabel('Temperatura')
plt.xlabel('Año')
plt.show()
```

De acuerdo con la gráfica:

      Respuesta: En tunja
